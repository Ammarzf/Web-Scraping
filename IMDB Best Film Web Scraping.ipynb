```python
import requests
from requests import get
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
headers = {'Accept-Language' : 'en-US, en;q=0.5'}
```

```python
url = 'https://www.imdb.com/search/title/?groups=top_1000&ref_=adv_prv'
result = requests.get(url, headers = headers)
```

```python
soup = bs4.BeautifulSoup(result.text, 'lxml')
```

```python
titles = []
years = []
times = []
imdb_ratings = []
metascores = []
votes = []
us_gross = []
```

```python
movie_div = soup.find_all('div', class_ = 'lister-item mode-advanced')
```

```python
movie_div
```

```python
for i in movie_div:
    title = i.h3.a.text
    titles.append(title)
    
    year = i.h3.find('span', class_= 'lister-item-year').text
    years.append(year)
    
    time = i.p.find('span', class_= 'runtime').text
    times.append(time)
    
    rating = float(i.strong.text)
    imdb_ratings.append(rating)
    
    score = i.find('span', class_= 'metascore').text if i.find('span', class_= 'metascore') else '-'
    metascores.append(score)
    
    nv = i.find_all('span', attrs = {'name' : 'nv'})
    
    vote = nv[0].text
    votes.append(vote)
    
    gross = nv[1].text if len(nv) > 1 else '-'
    us_gross.append(gross)    
```

```python
print(titles)
print(years)
print(times)
print(imdb_ratings)
print(metascores)
print(votes)
print(us_gross)
```

```python
movies = pd.DataFrame({
    'movie' : titles,
    'year' : years,
    'timeMin' : times,
    'imdb' : imdb_ratings,
    'metascore' : metascores,
    'votes' : votes,
    'us_grossMillions' : us_gross
})
```

```python
print(movies)
```

```python
print(movies.dtypes)
```

## Cleaning Data

```python
#remove parenthesis and convert object into integer data type
```

> ### Cleaning data:year

```python
movies['year'] = movies['year'].str.extract('(\d+)').astype(int)
```

```python
print(movies['year'])
```

> ### Cleaning data:time

```python
movies['timeMin'] = movies['timeMin'].str.extract('(\d+)').astype(int)
```

```python
print(movies['timeMin'])
```

> ### Cleaning data:votes

```python
movies['votes'] = movies['votes'].str.replace(',','').astype(int)
```

```python
movies['votes']
```

> ### Cleaning data:gross

```python
#Remove $ at leading and 'M' at trailing
movies['us_grossMillions'] = movies['us_grossMillions'].map(lambda x: x.lstrip('$').rstrip('M'))

movies['us_grossMillions'] = pd.to_numeric(movies['us_grossMillions'], errors = 'coerce')
```

```python
movies['us_grossMillions']
```

```python
print(movies)
print(movies.dtypes)
```

## Saving data to CSV

```python
movies.to_csv('movies.csv')
```
