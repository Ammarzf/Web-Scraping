```python
import requests
from requests import get
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

from time import sleep
from random import randint
```

```python
titles = []
years = []
times = []
imdb_ratings = []
metascores = []
votes = []
us_gross = []
```

```python
headers = {"Accept-Language": "en-US, en;q=0.5"}
```

```python
pages = np.arange(1,1001,50)
    
for page in pages: 
  
  page = requests.get("https://www.imdb.com/search/title/?groups=top_1000&start="+str(page)+"&ref_=adv_nxt", headers=headers)
  
  soup = BeautifulSoup(page.text, 'html.parser')
  
  movie_div = soup.find_all('div', class_='lister-item mode-advanced')
  
  sleep(randint(2,10))
```

## Scraping code

```python
for i in movie_div:
    
    title = i.h3.a.text
    titles.append(title)
    
    year = i.find('span', class_='lister-item-year').text
    years.append(year)
    
    time = i.find('span', class_='runtime').text
    times.append(time)
    
    rating = float(i.strong.text)
    imdb_ratings.append(rating)
    
    score = i.find('span', class_='metascore').text if i.find('span', class_='metascore') else '-'
    metascores.append(score)
    
    nv = i.find_all('span', attrs= {'name':'nv'})
    
    vote = nv[0].text
    votes.append(vote)
    
    gross = nv[1].text if len(nv) > 1 else '-'
    us_gross.append(gross)
```

```python
print(len(titles))
print(len(years))
print(len(times))
print(len(imdb_ratings))
print(len(metascores))
print(len(votes))
print(len(us_gross))
```

```python
imdb_movies = pd.DataFrame({
    'movie' : titles,
    'year' : years,
    'timeMin' : times,
    'imdb' : imdb_ratings,
    'metascore' : metascores,
    'votes' : votes,
    'us_grossMillions' : us_gross
})
```

```python
imdb_movies['year'] = imdb_movies['year'].str.extract('(\d+)').astype(int)

imdb_movies['timeMin'] = imdb_movies['timeMin'].str.extract('(\d+)').astype(int)

imdb_movies['votes'] = imdb_movies['votes'].str.replace(',', '').astype(int)

imdb_movies['metascore'] = imdb_movies['metascore'].str.extract('(\d+)')
imdb_movies['metascore'] = pd.to_numeric(imdb_movies['metascore'], errors='coerce')

imdb_movies['us_grossMillions'] = imdb_movies['us_grossMillions'].map(lambda x: x.lstrip('$').rstrip('M'))
imdb_movies['us_grossMillions'] = pd.to_numeric(imdb_movies['us_grossMillions'], errors='coerce')
```
